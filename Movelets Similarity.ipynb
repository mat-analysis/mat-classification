```python
import pandas as pd
movelets = pd.read_csv('./sample/results/hiper-vanessa/train.csv')
movelets.head()
```

```python
for c in movelets.columns:
    display(movelets[c].unique())
```

```python
movelets[movelets == 0] = 1
movelets[movelets == 2] = 0
movelets
```

```python
freqclass = movelets['label'].value_counts()
freqclass, 'Label 885:', freqclass[885] 
```

```python
maux = movelets.groupby('label').sum()
maux
```

```python
# Montando os dados estatsticos das movelets por classe:
import pandas as pd
stats = []
for c in movelets.columns[:-2]:
    line = {}
    contains = maux.index[maux[c]>=1.0].tolist()
    line = {
        'movelet': c,
        'origin_label': int(c.split('_CLASS')[-1]),
        'labels': contains,
#        'labels_prop': {x: maux[c][x] / freqclass[x] for x in contains},
#        'labels_count': {x: maux[c][x] for x in contains},
    }
    
    for l in labels:
        line['count_'+str(l)] = int(maux[c][l])
        line['prop_'+str(l)] = maux[c][l] / freqclass[l]
        
    stats.append(line)
    
pd.DataFrame(stats)
```

```python
# Agora vamos ver por classes que tem mais movelets em comum, uma tabela de similaridade:
#df_sim = 
df_sim = pd.DataFrame([[0]*len(labels)]*len(labels), columns=labels, index=labels)

for s in stats:
    for l in s['labels']:
        if s['origin_label'] in labels and l in labels:
            df_sim[s['origin_label']][l] += 1
    
df_sim
```
